```python
import pandas as pd

# example dataset
data = {
    'Weather': ['Sunny','Sunny','Rainy','Sunny','Rainy'],
    'Play':    ['No','No','Yes','Yes','Yes']
}

df = pd.DataFrame(data)

# function to calculate probabilities
def naive_bayes(feature_value, feature_name, target_value):
    df_target = df[df['Play'] == target_value]
    prob_feature = len(df_target[df_target[feature_name] == feature_value]) / len(df_target)
    prob_target = len(df_target) / len(df)
    return prob_feature * prob_target

# classify a new sample
P_yes = naive_bayes('Sunny', 'Weather', 'Yes')
P_no  = naive_bayes('Sunny', 'Weather', 'No')

print("P(Yes | Sunny):", P_yes)
print("P(No | Sunny):", P_no)

if P_yes > P_no:
    print("Prediction: Yes")
else:
    print("Prediction: No")
```
